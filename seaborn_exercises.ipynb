```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
# Use the iris database to answer the following quesitons:
```

```python
# 1. What does the distribution of petal lengths look like?
    
sns.displot(data = iris, x = "petal_length")
```

```python
# 2. Is there a correlation between petal length and petal width?
# R.- Yes, there is a stong correlation between both variables

sns.heatmap(iris.corr(), cmap = "Blues", annot = True, center = 0)
```

```python
# 3. Would it be reasonable to predict species based on sepal width and sepal length?
# R. This assumption has limited applicability

sns.displot(data = iris, x = "sepal_width", y = "sepal_length", hue = "species", height = 8.27, aspect = 11.7/8.27)
```

```python
# 4. Which features would be best used to predict species
# R. petal_width and petal_length

sns.pairplot(data = iris, hue = "species", height = 3)
```
